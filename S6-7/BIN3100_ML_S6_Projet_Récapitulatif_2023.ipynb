## **II) PRE-PROCESSING :**

##Objectif :  
- Transformer le data pour le mettre dans un format propice au machine learning

##Explications de vos choix :

```python
#Code pour votre pre-processing
from sklearn.model_selection import train_test_split
from sklearn.compose import make_column_transformer
from sklearn.model_selection import GridSearchCV
from sklearn.impute import SimpleImputer, KNNImputer
from sklearn.preprocessing import OneHotEncoder, LabelEncoder, OrdinalEncoder, LabelBinarizer
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import PolynomialFeatures
from sklearn.svm import LinearSVC
from sklearn.linear_model import SGDClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier

# Data et target
X = data.drop(['ID', 'Var_1', 'Segmentation'], axis = 1)
y = data['Segmentation']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 4)

ct = make_column_transformer(
    (OneHotEncoder(), ['Gender', 'Ever_Married', 'Profession']),
    (OrdinalEncoder(), ['Spending_Score']),
)

# Encoder les données
# X = OrdinalEncoder().fit_transform(X)

X = SimpleImputer(missing_values = np.nan, strategy = 'most_frequent').fit_transform(X)

params_randomforest = {
    'polynomialfeatures__degree': [2, 3, 4],
    'randomforestclassifier__max_depth': np.arange(1, 10)
}

model_randomforest = make_pipeline(ct, PolynomialFeatures(), RandomForestClassifier(random_state = 4))
grid_randomforest = GridSearchCV(model_randomforest, param_grid = params_randomforest, cv = 5)
grid_randomforest.fit(X_train, y_train)
print("Random forest")
print(grid_randomforest.best_score_)
print(grid_randomforest.best_params_)
print(grid_randomforest.score(X_test, y_test))
print()

params_kneighbors = {
    'polynomialfeatures__degree': [2, 3, 4],
    'kneighborsclassifier__n_neighbors': np.arange(1, 20)
}

model_kneighbors = make_pipeline(ct, PolynomialFeatures(), KNeighborsClassifier())
grid_kneighbors = GridSearchCV(model_kneighbors, param_grid = params_kneighbors, cv = 5)
grid_kneighbors.fit(X_train, y_train)
print("KNeighbors")
print(grid_kneighbors.best_score_)
print(grid_kneighbors.best_params_)
print(grid_kneighbors.score(X_test, y_test))
print()

params_sgd = {
    'polynomialfeatures__degree': [2, 3, 4],
    'sgdclassifier__penalty': ['l1', 'l2']
}

model_sgd = make_pipeline(ct, PolynomialFeatures(), SGDClassifier(random_state = 4))
grid_sgd = GridSearchCV(model_sgd, param_grid = params_sgd, cv = 5)
grid_sgd.fit(X_train, y_train)
print("SGD")
print(grid_sgd.best_score_)
print(grid_sgd.best_params_)
print(grid_sgd.score(X_test, y_test))
print()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.compose import make_column_transformer
from sklearn.model_selection import GridSearchCV
from sklearn.impute import SimpleImputer, KNNImputer
from sklearn.preprocessing import OneHotEncoder, LabelEncoder, OrdinalEncoder, LabelBinarizer
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import PolynomialFeatures
from sklearn.svm import LinearSVC
from sklearn.linear_model import SGDClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression

X = data.drop(['ID', 'Var_1', 'Segmentation'], axis = 1)
y = data['Segmentation']

# Nan
mode_ever_married = X['Ever_Married'].mode()[0]
X['Ever_Married'].fillna(mode_ever_married, inplace = True)
mode_graduated = X['Graduated'].mode()[0]
X['Graduated'].fillna(mode_graduated, inplace = True)
mode_profession = X['Profession'].mode()[0]
X['Profession'].fillna(mode_profession, inplace = True)
median_work_experience = X['Work_Experience'].median()
X['Work_Experience'].fillna(median_work_experience, inplace = True)
median_family_size = X['Family_Size'].median()
X['Family_Size'].fillna(median_family_size, inplace = True)

# Encodage variables binaires
label_encoder = LabelEncoder()
X['Gender'] = label_encoder.fit_transform(X['Gender'])
X['Ever_Married'] = label_encoder.fit_transform(X['Ever_Married'])
X['Graduated'] = label_encoder.fit_transform(X['Graduated'])

X['Spending_Score'] = label_encoder.fit_transform(X['Spending_Score'])

X = pd.get_dummies(X, columns = ['Profession'], prefix = ['Profession'])

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 4)



model = make_pipeline(PolynomialFeatures(), StandardScaler(), LogisticRegression(max_iter=1000))

params = {
    'polynomialfeatures__degree': [2, 3],
    'logisticregression__penalty': ['l1', 'l2', 'elasticnet', None],
    'logisticregression__solver': ['lbfgs', 'liblinear', 'newton-cg', 'newton-cholesky', 'sag', 'saga']
}

grid = GridSearchCV(model, param_grid = params, cv = 5)
grid.fit(X_train, y_train)
print(grid.best_score_)
print(grid.best_params_)
print(grid.score(X_test, y_test))
```

##**III) MODELLING**

##Objectif :  
- Développer un modèle de machine learning capable de répondre a l'objectif final.

##Explications de vos choix :

```python
#Code votre modelling
# Train test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 4)

# Params du gridsearch
params = {
    'polynomialfeatures__degree': [2, 3],
    'randomforestclassifier__max_depth': np.arange(1, 10)
}

# MIN MAX
model_min_max = make_pipeline(PolynomialFeatures(), MinMaxScaler(), RandomForestClassifier(random_state = 4))
grid_min_max = GridSearchCV(model_min_max, param_grid = params, cv = 5)
grid_min_max.fit(X_train, y_train)
print("MIN MAX")
print(grid_min_max.best_score_)
print(grid_min_max.best_params_)
print(grid_min_max.score(X_test, y_test))

# STANDARD
model_standard = make_pipeline(PolynomialFeatures(), StandardScaler(), RandomForestClassifier(random_state = 4))
grid_standard = GridSearchCV(model_standard, param_grid = params, cv = 5)
grid_standard.fit(X_train, y_train)
print("STANDARD")
print(grid_standard.best_score_)
print(grid_standard.best_params_)
print(grid_standard.score(X_test, y_test))

# ROBUST
model_robust = make_pipeline(PolynomialFeatures(), RobustScaler(), RandomForestClassifier(random_state = 4))
grid_robust = GridSearchCV(model_robust, param_grid = params, cv = 5)
grid_robust.fit(X_train, y_train)
print("ROBUST")
print(grid_robust.best_score_)
print(grid_robust.best_params_)
print(grid_robust.score(X_test, y_test))
```
