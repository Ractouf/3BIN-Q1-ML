## **II) PRE-PROCESSING :**

##Objectif :  
- Transformer le data pour le mettre dans un format propice au machine learning

##Explications de vos choix :
On commence par diviser notre dataset en données X et données y tout en retirant les colonnes qu'on a jugé inutiles dans notre analyse.

Ensuite on divise notre dataset en données de test et de train.

On a crée 2 column transformers afin d'éxecuter les encoders apres les imputers.

Ensuite on crée un column transformer pour les imputers, on a remarqué que si on mettait pas toutes les colonnes dans le simple imputer, le reste se faisait éliminer. On a donc mit toutes les colonnes dans le imputer

Ensuite on crée un autre column transformer pour les encoders. La colonne Gender et Profession sont encodés en utilisant le one hot, vu qu'elles ne sont pas comparables. Le reste des données str (Ever_Married, Graduated, Spending_Score) sont encodés avec un OrdinalEncoder car elles sont comparables.

Les imputer transforment nos données en tableaux numpy, c'est pourquoi dans le deuxième column transformer on a des indexe et non des noms de colonnes

```python
from sklearn.model_selection import train_test_split
from sklearn.compose import make_column_transformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder

X = data.drop(['ID', 'Var_1', 'Segmentation'], axis = 1)
y = data['Segmentation']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25)

ct_imputer = make_column_transformer(
    # (SimpleImputer(missing_values = np.nan, strategy = 'most_frequent'), ['Work_Experience', 'Family_Size']),
    # (SimpleImputer(missing_values = np.nan, strategy = 'most_frequent'), ['Ever_Married', 'Graduated', 'Profession'])
    (SimpleImputer(missing_values = np.nan, strategy = 'most_frequent'), ['Gender', 'Ever_Married', 'Age', 'Graduated', 'Profession', 'Work_Experience', 'Spending_Score', 'Family_Size'])
)

ct_encoder = make_column_transformer(
    (OrdinalEncoder(), [1, 3, 6]),
    (OneHotEncoder(sparse_output = False), [0, 4])
)
```

##**III) MODELLING**

##Objectif :  
- Développer un modèle de machine learning capable de répondre a l'objectif final.

##Explications de vos choix :
Avec tout notre préprocessing de fini, on place tout dans notre pipeline suivi d'un PolynomialFeatures et de notre estimator choisi.
Nous n'avons pas de scaler car le RandomForestClassifier n'en a pas besoin.

Ensuite on effectue un grid search afin de trouver les meilleurs hyperparamètres et le meilleur modèle

```python
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import PolynomialFeatures
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV

model = make_pipeline(ct_imputer, ct_encoder, PolynomialFeatures(), RandomForestClassifier())

params = {
    'polynomialfeatures__degree': [2, 3],
    'randomforestclassifier__max_depth': np.arange(1, 10)
}

grid = GridSearchCV(model, param_grid = params, cv = 5)
grid.fit(X_train, y_train)
print(f'# {grid.best_params_}')
print(f'# BEST SCORE {grid.best_score_}')
print(f'# TRAIN {grid.score(X_train, y_train)}')
print(f'# TEST {grid.score(X_test, y_test)}')

model_final = grid.best_estimator_
```

# Teste de notre modèle sur l'accuracy

Comme dit dans l'analyse nous allons tester notre modèle sur son accuracy.
Le modèle obtient des résultats assez similaires mais le train a un meilleur accuracy, ce qui est évidemment normal.

Avec l'accuracy on obtient un score qui est meilleur que le hasard (qui est de 25% vu qu'on a 4 target).

```python
from sklearn.metrics import accuracy_score

predicted_labels_train = model_final.predict(X_train)
predicted_labels_test = model_final.predict(X_test)

# Calculate accuracy
accuracy_train = accuracy_score(y_train, predicted_labels_train)
print(f"Accuracy train: {accuracy_train}")
accuracy_test = accuracy_score(y_test, predicted_labels_test)
print(f"Accuracy test: {accuracy_test}")
```

# Learning curve
Sur le graphique ci dessous on peut voir qu'on n'a pas de overfitting et que le modèle pourrait encore devenir meilleur si il disposait de plus de données.

```python
from sklearn.model_selection import learning_curve

N, train_score, val_score = learning_curve(model_final, X_train, y_train, train_sizes = np.linspace(0.1, 1.0, 10), cv = 5)

plt.plot(N, train_score.mean(axis = 1), label = 'train')
plt.plot(N, val_score.mean(axis = 1), label = 'val')
```
